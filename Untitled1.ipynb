任务：
加载数据：读取 employees.csv 文件到一个 DataFrame 中。

处理缺失值：
对于 age 列的缺失值，填充为该列的平均值。

对于 gender 列的缺失值，用 U (未知) 填充。

对于 salary 列的缺失值，填充为该列的中位数。

对于 department 列的缺失值，填充为 Unknown。

对于 hire_date 列的缺失值，填充为 2020-01-01。

数据类型转换：将 hire_date 列转换为日期格式。

删除重复数据：确保 id 列没有重复值，如果有，保留最后出现的那一行。
数据校正：

将 gender 列的值转换为统一的大写格式（例如，将所有 f 转换为 F，m 转换为 M）。

数据筛选：筛选出 age 在 30 岁以上且 salary 大于 50000 的员工。

保存清洗后的数据：将清洗后的数据保存为新的 CSV 文件 cleaned_employees.csv。

```python
# 处理缺失值： 对于 age 列的缺失值，填充为该列的平均值。
df['age'].isnull()
df['age'] = df['age'].fillna(df['age'].mean()).astype(int)
df
```

```python
# 对于 gender 列的缺失值，用 U (未知) 填充
df['gender'] = df['gender'].fillna("U")
df
```

```python
# 对于 salary 列的缺失值，填充为该列的中位数。
df['salary'] = df['salary'].fillna(df['salary'].median())
df
```

```python
# 对于 department 列的缺失值，填充为 Unknown。
df['department'] = df['department'].fillna("Unkonwn")
df
# 对于 hire_date 列的缺失值，填充为 2020-01-01
df['hire_date'] = df['hire_date'].fillna("2020-01-01")
df
# 将hire_date列改为datetime格式
df['hire_date'] = pd.to_datetime(df['hire_date'])
df.info()
```

```python
# 删除重复数据：确保 id 列没有重复值，如果有，保留最后出现的那一行。
df['id'].duplicated()
df = df.drop_duplicates(subset=['id'],keep='last')
df
```

```python
# 将 gender 列的值转换为统一的大写格式（例如，将所有 f 转换为 F，m 转换为 M）
df['gender'] = df['gender'].str.upper()
df
```

```python
# 数据筛选：筛选出 age 在 30 岁以上且 salary 大于 50000 的员工。
df.query('age>30 and salary>5000 ')
# df[df['age']>30 & df['salary']>5000]
```
